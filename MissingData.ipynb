```python
import numpy as np
import pandas as pd
```

Finding Missing Data

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [np.nan, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.isna().any()
```

Removing Missing Data

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df2 = pd.DataFrame(data)
```

```python
df2
```

```python
df2.dropna()
```

```python
df2
```

```python
df2.dropna(thresh=3)
```

```python
df2.dropna(thresh=4)
```

```python
df2
```

```python
df2.fillna(0)
```

```python
values = {'A': 0, 'B': 15, 'C': 24, 'D': 76}
df2.fillna(value=values)
```

```python
df2.fillna(df.mean())
```
